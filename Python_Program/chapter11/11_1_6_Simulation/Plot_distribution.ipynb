```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



df = pd.read_csv('between.csv')
df
```

```python
df = df.dropna()
df
```

```python
df.plot.bar(x='Between_time', y='Frequency')
plt.xticks(rotation=30, horizontalalignment="center")
plt.savefig('freq_between.eps', format='eps')
```

```python
df.plot.bar(x='Between_time', y='Cumulative_Probability')
plt.xticks(rotation=30, horizontalalignment="center")
plt.savefig('com_between.eps', format='eps')
```

```python
b = df['slope_b']*df['Cumulative_Probability'] + df['intercept_b']
```

```python
df.dropna()
df['Between'] = b
df
```

```python
df.plot(x='Cumulative_Probability',y='Between')
plt.savefig('com_between2.eps', format='eps')
```
