```python
import numpy as np
import matplotlib.pyplot as plt

Ds = 7
D0 = 6
B0 = 20
ka = 0.2
kd = 0.1
ubar = 2.0

x = np.linspace(0, 200, 100)
y = Ds - (Ds-D0)*np.exp(-ka*x/ubar) - ((kd*B0)/(ka-kd))*(np.exp(-kd*x/ubar)-np.exp(-ka*x/ubar))

plt.plot(x, y, '-', linewidth=2)
plt.xlabel('x')
plt.ylabel('D')
plt.savefig('DO-sag.eps', format='eps')
```

```python
x = np.linspace(0, 200, 100)
B = B0*np.exp(-kd*x/ubar)

plt.plot(x, B, '-', linewidth=2)
plt.xlabel('x')
plt.ylabel('B')
plt.savefig('BOD.eps', format='eps')
```

```python
x = np.linspace(0, 200, 100)
Dd = (Ds-D0)*np.exp(-ka*x/ubar) + ((kd*B0)/(ka-kd))*(np.exp(-kd*x/ubar)-np.exp(-ka*x/ubar))

plt.plot(x, Dd, '-', linewidth=2)
plt.xlabel('x')
plt.ylabel('Dd')
plt.savefig('DO-deficit.eps', format='eps')
```
