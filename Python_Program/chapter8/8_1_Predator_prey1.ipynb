```python
import matplotlib.pyplot as plt

def euler(fs, t0, y0s, dt):
    ns = len(y0s)
    for i in range(ns):
      y0s[i] =  y0s[i] + fs[i](t0, y0s)*dt 
    return y0s 
    
def sys_ode(fs, t0, y0s, dt, tn):
  nt = int((tn-t0)/dt)  
  for i in range(nt):
    y0s = euler(fs, t0, y0s, dt) 
    t0  = t0 + dt
    plt.plot(t0,y0s[0],'bo')
    plt.plot(t0,y0s[1],'r+')
    plt.xlabel('t')
    plt.ylabel('predator-prey')
  return y0s

k1 = 0.5
k2 = 0.02
k3 = 0.004
k4 = 0.4

dt = 0.01 
tn = 50
t0 = 0
x0 = 20
y0 = 40

fs = [
    lambda t, ys : k1*ys[0] - k2*ys[0]*ys[1],
    lambda t, ys : k3*ys[0]*ys[1] - k4*ys[1]
] 

ys = sys_ode(fs, t0, [x0, y0], dt, tn)
print(ys)

```
