```python
# Example 1 - fit many parameters

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
td = [0, 1, 2, 3, 4, 5, 6, 7]
yd = [12, 20, 32, 54, 78, 92, 112, 125]
```

```python
plt.plot(td, yd, 'ro')
plt.axis([0, 8, 0, 140])
plt.show()
```

```python
def f(t,A, B, C):
    """ Fit function y=f(t, A, B, C) with parameters A, B, C """
    y = A/(1+B*np.exp(-C*t))
    return y
```

```python
# call curve fit function
popt , pcov = curve_fit (f, td, yd)
print('Fitted parameters',popt)
A = popt[0]
B = popt[1]
C = popt[2]
print('A =', A, 'B =',B, 'C =', C)
```

```python
tf = np.linspace(0, 8)

yf = A/(1+B*np.exp(-C*tf))

plt.plot(tf, yf, '--', linewidth=2)
plt.xlabel('t')
plt.ylabel('y')
plt.legend(['Logistic model'])
plt.show()
```

```python
plt.plot(tf, yf, '--', linewidth=2)
plt.plot(td, yd, 'ro')
plt.xlabel('t')
plt.ylabel('y')
plt.legend(['Logistic model','Data'])
plt.show()
```
